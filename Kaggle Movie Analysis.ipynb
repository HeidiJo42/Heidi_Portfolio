```python
# Load packages 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Load Dataset from Kaggle
# Link: https://www.kaggle.com/danielgrijalvas/movies

df = pd.read_csv(r'C:\Users\hjlse\OneDrive\Programming Training\GitHub Docs\Python Movies Project\Kaggle Data\movies.csv')
```

```python
# Look at the data load
df.head()
```

```python
# Check for missing data 
# Example of a for loop

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change $$ variables into int

# df['budget'] = df['budget'].astype('int64')
# df['gross'] = df['gross'].astype('int')
# int & int64 through errors: 
# fixed using float type

df["gross"] = df["gross"].astype(str).astype(float)
df["budget"] = df["budget"].astype(str).astype(float)

print(df.dtypes)
```

```python
# Order our Data by gross

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_factor = df

for col_name in df_factor.columns:
    if(df_factor[col_name].dtype == 'object'): 
        df_factor[col_name] = df_factor[col_name].astype('category')
        df_factor[col_name] = df_factor[col_name].cat.codes
        
df_factor
```

```python
correlation_matrix = df_factor.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_factor.corr()
```

```python
corr_matrix = df_factor.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sort_pairs = corr_pairs.sort_values()

sort_pairs
```

```python
high_corr = sort_pairs[(sort_pairs > 0.4)]
high_corr
```

```python


# Using factorize - 
# this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

```

```python
# correlation matrix using lambda function
# method 2

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()

```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Most profitable years per company

CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Top Grossing companies

CompanyGrossSum = df.groupby(['company'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
